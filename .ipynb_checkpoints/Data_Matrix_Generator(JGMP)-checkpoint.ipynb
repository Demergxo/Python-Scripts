```python
#pip install pylibdmtx
#pip install pillow


from pylibdmtx.pylibdmtx import encode
from pylibdmtx.pylibdmtx import decode
from PIL import Image, ImageFont, ImageDraw 
import os
import cv2

letter= "P"
location_string = "{} .".format(letter)
directory = os.getcwd()

for i in range(1,71):
    
    zero_fill = str(str(i).zfill(3))
    iterate_string = location_string + zero_fill
    

    encoded = encode(iterate_string.encode('utf8'))#, scheme="Ascii", size = "32x32")
    #encoded = encode("R08.04110C".encode('utf8'))
    img = Image.frombytes('RGB', (encoded.width, encoded.height), encoded.pixels)
    
    barcode="{}.{}.png".format(letter, zero_fill)
    dbarcode = directory +"\\"+barcode
    
    img.save(dbarcode)
    
    print(decode(Image.open(dbarcode)))
    
```

# **CREACIÓN DE LAS FUNCIONES**

```python
# obtener hora y dia

def actual_date():
    now = datetime.now()
    date_now = now.date()
   
    return date_now

def actual_hour():
    now = datetime.now()
    hour_now = now.time()

    return hour_now

#seleccionar archivo correcto en descargas

def select_file():

    pattern = r"^(-[\w .]*-)([\d]{8})"
    pattern_date = r"^([\d]*)-([\d]*)-([\d]*)"
        
    date_now = str(actual_date())
    control_date = re.sub(pattern_date, r"\1\2\3", date_now)
    

    directory = "C:\\Users\jgmeras\\Downloads\\"

    for name in os.listdir(directory):
        fullname = os.path.join(directory, name)
        string = re.sub(pattern, r"\2", name)
        if not os.path.isdir(fullname):

            sub_result = re.search(pattern_date, string)

            if any(x in sub_result[0] for x in control_date):
                return name
            else:
                print("REVISAR")



#cambiar archivo de directorio

def change_path():
    path_source = "C:\\Users\\jgmeras\\Downloads\\" + select_file()
    path_destination = "C:\\Users\\jgmeras\\Documents\\Python Scripts\\" + select_file()
    print(path_destination)

    os.replace(path_source, path_destination)



#seleccionar archivo en directorio actual


def select_file2():
    pattern = r"\\([\d]{8})"
    pattern_date = r"^([\d]*)-([\d]*)-([\d]*)"
    #pattern_third = r"(\[\]\')"

    date_now = str(actual_date())
    control_date = re.sub(pattern_date, r"\1\2\3", date_now)
    
    dir = os.getcwd()
    
    for name in os.listdir(dir):
        fullname = os.path.join(dir, name)
        
        if not os.path.isdir(fullname):
            
            sub_result = re.search(pattern, fullname)
            
            #result = re.sub(pattern_third,  r"", str(sub_result))
            if any(x in sub_result[0] for x in control_date):
                return name
            else:
                print("REVISAR")
```

# **CONECTAR AL SERVIDOR SMTP Y ENVIAR EL ARCHIVO**

```python
msg = MIMEMultipart()

files =  select_file2()

user = 'javier.garcia-meras-palacios@gxo.com'
msg['From'] = user
msg['To'] = user
msg['Subject'] = "Inventario Diario"

part = MIMEBase('application', "octet-stream")
with open(files, 'rb') as file:
    part.set_payload(file.read())
encoders.encode_base64(part)
part.add_header('Content-Disposition',
                'attachment; filename={}'.format(file))
msg.attach(part)



server = smtplib.SMTP('10.19.30.20:25' )
server.set_debuglevel(1)


server.ehlo()
server.starttls()
server.ehlo()



try:
    server.sendmail(
        msg['From'],
        msg['To'],
        msg.as_string())
 
    server.quit()
 
    print ("successfully sent email to %s:" % (msg['To']))
except:
    print("unable to connect")

driver.close()
```

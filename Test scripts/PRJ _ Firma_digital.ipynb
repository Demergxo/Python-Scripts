# Generación de claves

```python
from Cypto import Random
from Crypto.publicKey import RSA

# Generador de numeros aleatorios
random_generator = Random.new().read
# Generación de algoritmo RSA
rsa = RSA.generate(1024, random_generator)
# Generación de claves maestras
private_pem = rsa.exportKey()
public_pem = rsa.publickKey().exportKey

with open('ClavePrivadaGXOElux.pem', 'wb') as f:
    f.write(private_pem)

with open ('ClavePublicaGXOElux.pem', 'wb') as f:
    f.write(public_pem)

print("Se han generado las dos claves:")
print(("Privada: {}").format(private_pem))
print(("Publica: {}").format(public_pem))
```

# Firmar un documento

```python
from Crypto.Hash import SHA
from Crypto.Signature import PKCS1_v1_5 as Signature_pkcs1_v1_5
from Crypto.PublicKey import RSA
import base64

# Usar clave para firmar el archivo

def firma(message):
    with open ('ClavePrivadaGXOElux.pem') as f:
        key= f.read()
        rsakey = RSA.importKey(key)
        signer = Signature_pkcs1_v1_5(rsakey)

        #Cifrado HASH
        
        digest = SHA.new()
        
        #Firmamos
        
        digest.update(message)
        print("Contenido del documento: ", message)
        print("Calculamos hash del documento: ", digest.hexdigest())

        sign = signer.sign(digest)
        signature = base64.b64encode(sign)

with open ("notas.txt", "r") as file:
    message=file.read()

message = message.encode()
signature = firma(message)



```

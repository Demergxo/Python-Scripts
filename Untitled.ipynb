```python
def busqueda_binaria_recursiva(lista, key, left, right):
    if left > right:
        return -1
    middlepos = (left + right) // 2
    middle = lista[middlepos]
    if middle == key:
        return middlepos
    if key < middle:
        return (middle - 1)
    else:
        return(middle + 1)
```

```python
from datetime import datetime
import time

def actual_date():
    now = datetime.now()
    date_now = now.date()
   
    return date_now

def actual_hour():
    now = datetime.now()
    hour_now = now.time()

    return hour_now

date = actual_date()
hour1 = actual_hour()
time1 = time.time()

print(("HORA: {} FECHA: {}").format(hour1, date))
time.sleep(30)

hour2 = actual_hour()
time2 = time.time()

print(("HORA: {} FECHA: {}").format(hour2, date))


time_interval = time2 - time1

print(("Diferencia: {}").format(time_interval))
```

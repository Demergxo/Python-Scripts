```python
#pip install pylibdmtx
#pip install pillow


from pylibdmtx.pylibdmtx import encode
from pylibdmtx.pylibdmtx import decode
from PIL import Image, ImageFont, ImageDraw 
import os
import cv2
import easygui as eg


        
```

```python
ubications= ["PRO" ]
directoy_name = eg.enterbox(msg="Nombre del Directorio donde se va a guardar:", title="Control", default="", strip=True, image=None)

directory = os.getcwd() + ("\\{}\\").format(directoy_name)

try:
    os.stat(directory)
except:
    os.mkdir(directory)


for loc in ubications:
    for i in range(0, 35):
        
        zero_fill = str(i).zfill(3)
        iterate_string = loc+" ." + zero_fill
        

        encoded = encode(iterate_string.encode('utf8'))#, scheme="Ascii", size = "32x32")
        img = Image.frombytes('RGB', (encoded.width, encoded.height), encoded.pixels)
        
        barcode="{}.{}.png".format(loc, zero_fill)
        dbarcode = directory + barcode
        
        img.save(dbarcode)
        
        print(decode(Image.open(dbarcode)))
```

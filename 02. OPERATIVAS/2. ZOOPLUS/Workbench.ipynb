```python
import tkinter as tk
from tkinterdnd2 import TkinterDnD, DND_FILES  # Necesitarás instalar tkinterdnd2

def archivo_arrastrado(event):
    # Obtener la ruta del archivo arrastrado
    ruta_archivo = event.data.strip("{}")
    # Actualizar el contenido del Entry con la ruta
    entrada_ruta.delete(0, tk.END)
    entrada_ruta.insert(0, ruta_archivo)

# Crear la ventana principal
root = TkinterDnD.Tk()  # Usar TkinterDnD para habilitar arrastrar y soltar
root.title("Arrastra un archivo aquí")
root.geometry("500x200")

# Etiqueta para indicar "Ruta"
label_ruta = tk.Label(root, text="Ruta del archivo:", font=("Arial", 12))
label_ruta.pack(pady=10)

# Zona de texto para mostrar y actualizar la ruta
entrada_ruta = tk.Entry(root, font=("Arial", 12), width=60, state="normal")
entrada_ruta.pack(pady=10)

# Crear una zona designada para arrastrar archivos
zona_arrastre = tk.Label(root, text="Arrastra un archivo aquí", bg="lightblue", width=60, height=5, relief="groove")
zona_arrastre.pack(pady=20)

# Asociar el evento de arrastrar y soltar
zona_arrastre.drop_target_register(DND_FILES)
zona_arrastre.dnd_bind('<<Drop>>', archivo_arrastrado)

# Iniciar el loop de la interfaz
root.mainloop()
```

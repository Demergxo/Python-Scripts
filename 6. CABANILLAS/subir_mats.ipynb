```python
from selenium.webdriver import Chrome
from selenium import webdriver
#from selenium.webdriver.support.ui import Select
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait as wait

import os
import openpyxl
import time
```

```python
#Incluir la localizacion del fichero excel
directory = os.getcwd()
path = os.path.join(directory, "Libro1.xlsx")
#print(directory)

# Se abre el workbook y la hoja seleccionados (donde se guardo el documento por ultima vez)
wb_obj = openpyxl.load_workbook(path)
sheet_obj = wb_obj.active

# Obtener el numero de filas y columnas usadas
row_limit = sheet_obj.max_row # type: ignore
column_limit = sheet_obj.max_column # type: ignore
  
print("Total de filas:", row_limit)
print("Total de columnas:", column_limit)

#Introducir el numero de columna en la que se encuentra la matricula (A=1, B=2, C=3...)
plate_col = 1

#Introducir el numero de la fila en la que se encuentra el primer valor 
first_row = 2

#Introducir el numero de columna en la que se encuentra el nombre
name_col = 2
```

```python
# ****** URL DE PRODUCCION ******
url = "http://10.59.6.7:4002/admin"
 
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

driver_options = Options()
driver_options.add_argument(headers)
driver_options.add_argument('--no-sandbox')

driver = webdriver.Chrome()
driver.maximize_window()
driver.get(url)
```

```python
from django.conf.urls import url
print("[+] INICIANDO ACTUALIZACIÓN DE MATRICULAS")
print("*"*32+"\n")
for i in range(first_row, row_limit + 1): #type: ignore
    cell_obj_plate = sheet_obj.cell(row = i, column = plate_col) # type: ignore
    plate = str(cell_obj_plate.value)
    cell_obj_name = sheet_obj.cell(row = i, column = name_col) # type: ignore
    name = str(cell_obj_name.value)
    
    try:
        driver.find_element("xpath", "/html/body/div/section/div/div/div[1]/form/div/div[1]/input").send_keys(plate)
        driver.find_element("xpath", "/html/body/div/section/div/div/div[1]/form/div/div[2]/input").send_keys(name)
        driver.find_element("xpath", "/html/body/div/section/div/div/div[1]/form/div/div[3]/input").click()
        print(f"vuelta: {i}, nombre: {name}")
    except:
        print(f"ruptura en línea {i+1}, nombre : {name}, matricula: {plate}")
        break
    #print(f"vuelta: {i}")
    time.sleep(3)
    print("*"*32+"\n")
    print("[+] ACTUALIZACIÓN DE MATRICULAS FINALIZADA")
    driver.close()
```

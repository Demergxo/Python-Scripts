```python
import tensorflow as tf
import numpy as np

#F = C *1.8+32
```

```python
celsius = np.array([-40, 10, 0, 8, 15, 22, 38, 45, 60, 80], dtype=float)
fahreneit = np.array([-40, 14, 32, 46.4 , 59, 72, 100.4, 113, 140, 176], dtype=float)
```

```python
#capa = tf.keras.layers.Dense(units=1, input_shape=[1])
#modelo = tf.keras.Sequential([capa])

oculta1 = tf.keras.layers.Dense(units=7, input_shape=[1])
oculta2 = tf.keras.layers.Dense(units=7)
oculta3 = tf.keras.layers.Dense(units=5)
oculta4 = tf.keras.layers.Dense(units=3)
salida = tf.keras.layers.Dense(units=1)
modelo = tf.keras.Sequential([oculta1, oculta2, oculta3, oculta4, salida])
```

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss = 'mean_squared_error'
)
```

```python
print("[+] Comenzando entrenamiento...")
historial = modelo.fit(celsius, fahreneit, epochs=1000, verbose=False)
print("[+] Modelo entrenando")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de perdida")
plt.plot(historial.history["loss"])
```

```python
print("[+] Predicción")
entrada_prediccion = np.array([120.0])
resultado = modelo.predict(entrada_prediccion)
print(" El resultado es {} F".format(resultado))
```
